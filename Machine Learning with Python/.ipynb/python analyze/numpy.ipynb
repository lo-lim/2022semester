#Numpy Library

```python
import numpy as np
v1=np.array([1,2,3,4])
print(v1)
```

```python
v1=np.arange(10)
print(v1)
```

```python
v2=np.arange(1,10,2, dtype=int)
v3=np.arange(3.5,10.5,2, dtype=float)
print(v2)
print(v3)
```

```python
v4=np.arange(1,10,2)**2
print(v4)
```

```python
v5=np.arange(1,10,2)**3
print(v5)
```

```python
v1=np.arange(12)
print(v1)
v2=v1.reshape(2,6)
print(v2)
v3=v1.reshape(2,6, order='F')
print(v3)
```

```python
v1=np.arange(1,5).reshape(2,2)
print(v1)
print(np.add(v1,v1))
print(np.subtract(v1,v1))
print(np.multiply(v1,v1))
print(np.dot(v1,v1))
```

```python
v1=np.arange(12)
print(v1)
v2=v1.reshape(2,3,2)
v3=v1.reshape(2,3,2,order='F')
print(v2)
print(v3)
```

```python
v3=np.arange(3.5, 10.5, 2, dtype=float)
print(v3)
v4=np.arange(1,5).reshape(2,2)
print(v4)
```

```python
v1=np.arange(12)
print(v1)
v2=v1.reshape(2,3,2, order='F')
print(v2)
```

```python
v3=np.arange(3.5, 10.5, 2, dtype=float)
print(v3)
v4=np.arange(1,5).reshape(2,2)
print(v4)
print(np.amax(v3))
print(np.amin(v4))
print(v3.dtype)
print(v4.shape)
```
