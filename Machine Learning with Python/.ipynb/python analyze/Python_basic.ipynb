## while문

```python
i=0
sum=0
while i<11:
    sum=sum+i
    print(i, sum)
    i=i+1
sum
```

```python
food=["apple", "banana", "bread"]
i=0
while i < len(food):
    print(food[i])
    i=i+1
```
